```python
# from selenium import webdriver
import time
import pandas as pd
from bs4 import BeautifulSoup
import re
import requests
import datetime
import pyodbc
import sqlite3



curtime=datetime.datetime.now()
curtime1=str(curtime.date())+str(curtime.time())[:2]+str(curtime.time())[3:5]+str(curtime.time())[6:8]
print('kkkkkkkkkkkkkkkkkkkkkkkkk',curtime1)
compnumber=1
n=1
laststate={}
a=requests.get('http://www.tsetmc.com/tsev2/data/MarketWatchInit.aspx?h=0&r=0').text
b=a.split(';')
c=[i.split(',') for i in b if len(i.split(','))==23]
d={i[0]:i[1:] for i in c}
```

```python
mdf=pd.DataFrame()
for compid in list(d.keys()):

    cid=str(compid)
    compnam=d[cid][2]
    compnem=d[cid][1]
    print('{} started. it is {} comp from {} companies'.format(compnam,compnumber,len(d)))
    compnumber+=1
    site = requests.get('http://members.tsetmc.com/tsev2/data/InstTradeHistory.aspx?i=' + str(compid) + '&Top=999999&A=0')
    print(site)
    hisorylist = site.text.split(';')
    print(len(hisorylist))
    hisn = 1
    for hisitem in hisorylist:
        dhisitem = hisitem.split('@')

        if (len(dhisitem)) > 9:
            tarik = dhisitem[0]
            if str(compid) in laststate.keys():
                if int(tarik)<=int(laststate[str(compid)]):
                    print(compnem,'compdata was exist----------->->->-break')
                    break
            maxprice = dhisitem[1]
            minprice = dhisitem[2]
            finishprice = dhisitem[3]
            lastprice = dhisitem[4]
            avalinprice = dhisitem[5]
            dirozprice = dhisitem[6]
            tradevalue = dhisitem[7]
            tradevol = dhisitem[8]
            tedad = dhisitem[9]
            curtime = curtime1
            
            cmdf=pd.DataFrame([str(n), str(tarik), str(compnam), str(compnem), str(maxprice), str(minprice),
                str(finishprice), str(lastprice), str(avalinprice), str(dirozprice), str(tradevalue), str(tradevol),
                str(tedad), str(curtime)])
            mdf=mdf.append(cmdf.T)
            n = n + 1
            print(hisn, len(hisorylist), 'compnumber', compnumber, int(100 * hisn / len(hisorylist)))
        hisn += 1
    print(len(hisorylist))
```

```python
mdf.columns=['id','tarik','cname','cnemad','maxprice','minprice','finishprice','lastprice','avalinprice','dirozprice','tradevalue','tradevol','tedad','curtime']
```
